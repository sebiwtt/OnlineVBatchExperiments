```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_cpu = pd.read_csv("./ARIMA_cpu_usage.csv")
df_ram = pd.read_csv("./ARIMA_ram_usage.csv")
```

```python
fig, axs = plt.subplots(1, 1, figsize=(15, 6))
axs.plot(df_cpu.value, label = "CPU")
axs.legend(fontsize=20)
plt.yticks(fontsize=10)
plt.xticks([])
plt.ylabel('CPU Usage [%]')
plt.xlabel('Time')
plt.tight_layout()
plt.savefig('./GeneratedPlots/ARIMACPU.png') 
plt.show()

fig, axs = plt.subplots(1, 1, figsize=(15, 6))
axs.plot(df_ram.value, label = "RAM")
axs.legend(fontsize=20)
plt.yticks(fontsize=10)
plt.xticks([])
plt.ylabel('RAM Usage [%]')
plt.xlabel('Time')
plt.tight_layout()
plt.savefig('./GeneratedPlots/ARIMARAM.png') 
plt.show()
```
