```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_cpu_arima = pd.read_csv("ARIMA_cpu_usage.csv")
df_cpu_pad = pd.read_csv("PAD_cpu_usage.csv")
df_cpu_prophet = pd.read_csv("Prophet_cpu_usage.csv")

df_ram_arima = pd.read_csv("ARIMA_ram_usage.csv")
df_ram_pad = pd.read_csv("PAD_ram_usage.csv")
df_ram_prophet = pd.read_csv("Prophet_ram_usage.csv")
```

```python
fig, axs = plt.subplots(1, 1, figsize=(15, 6))
axs.plot(df_cpu_arima.value[:19], label = "SARIMA")
axs.plot(df_cpu_pad.value[:19], label = "PAD")
axs.plot(df_cpu_prophet.value[:19], label = "Prophet")
axs.legend(fontsize=20)
plt.yticks(fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=12) 
plt.xticks([])
plt.ylabel('CPU Usage [%]',fontsize=14)
plt.xlabel('Time',fontsize=14)
plt.tight_layout()
plt.savefig('./CPU.png') 
plt.show()

fig, axs = plt.subplots(1, 1, figsize=(15, 6))
axs.plot(df_ram_arima.value[:32], label = "SARIMA")
axs.plot(df_ram_pad.value[:32], label = "PAD")
axs.plot(df_ram_prophet.value[:32], label = "Prophet")
axs.set_ylim(-0.0, 60.0) 
axs.legend(fontsize=20)
plt.yticks(fontsize=10)
plt.xticks([])
plt.tick_params(axis='both', which='major', labelsize=12) 
plt.ylabel('RAM Usage [%]',fontsize=14)
plt.xlabel('Time',fontsize=14)
plt.tight_layout()
plt.savefig('./RAM.png') 
plt.show()
```
